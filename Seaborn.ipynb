```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# loading dataset
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.shape
```

```python
# creating lineplot between timepoint nad signal
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
#2.Grouping data with hue
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
# 3. Adding Styles 
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
# 4. Adding Markers
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

```python
#importing required library
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# loading diamonds dataset
diamonds = pd.read_csv('diamonds.csv')
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

```python
# Assigning color in DistPLot
sns.distplot(diamonds['price'],color="red")
plt.show()
```

```python
sns.distplot(diamonds['price'],color="green",bins=5,kde=False)
plt.show()
```

```python
# Vertical Distplot
sns.distplot(diamonds['price'],color='blue',kde=False,bins=5,vertical=True)
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
# joint Plot
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=iris)
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',color='olive',data=iris)
plt.show()
```

```python
sns.jointplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,color='brown',kind='reg')
plt.show()
```

```python
# BoxPlot
churn = pd.read_csv('Churn.csv')
churn.head()
```

```python
sns.boxplot(x='Churn',y='tenure',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn)
plt.show()
```

```python
sns.boxplot(x='InternetService',y='MonthlyCharges',data=churn,palette="Set1")
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,linewidth=3)
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,order=['Two year','One year','Month-to-month'])
plt.show()
```

```python
sns.boxplot(x='Contract',y='tenure',data=churn,hue='PaymentMethod')
plt.show()
```
