```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 1. Line Plot
x=np.arange(1,11)
y=3*x
```

```python
plt.plot(x,y)
plt.show()
```

```python
#2. Adding Title and Labels
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-aixs")
plt.ylabel("y-axis")
plt.show()
```

```python
# 3. Changing Line Aesthetics
```

```python
plt.plot(x,y,color='r',linestyle=":",linewidth=4)
plt.title("Line Plot")
plt.xlabel("X-aixs")
plt.ylabel("y-axis")
plt.show()
```

```python
# 4. Adding two Lines in same plot.
```

```python
x=np.arange(1,11)
y=x*2
z=x*3
plt.plot(x,y,color='r',linestyle=":",linewidth=4)
plt.plot(x,z,color='y',linestyle="--",linewidth=2)
plt.title("Line Plot")
plt.xlabel("X-aixs")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#4.1 Adding labels inside plot
```

```python
year = [1972, 1982, 1992, 2002, 2010] 
e_india = [100.6, 158.61, 305.54, 394.96, 724.79] 
e_america = [10.5, 25.21, 58.65, 119.27, 274.87] 

# plotting of x-axis(year) and  y-axis(power consumption) 
plt.plot(year, e_india, color ='orange', label ='India') 
plt.plot(year, e_america, color ='g', label ='america') 

# naming of x-axis and y-axis 
plt.xlabel('Years') 
plt.ylabel('Power consumption in kWh') 

# naming the title of the plot 
plt.title('Electricity consumption per capita\ of India and America') 
plt.legend() 
plt.show()
```

```python
# 5. Adding Subplots
```

```python
x=np.arange(1,11)
y=x*2
z=x*3
plt.subplot(1,2,1)
plt.plot(x,y,color='b',linestyle=":",linewidth=4)

plt.subplot(1,2,2)
plt.plot(x,z,color='g',linestyle="--",linewidth=2)

plt.show()
```

```python
x=np.arange(1,11)
y=x*2
a=np.arange(100,1001)
b=a*0.5
plt.subplot(2,1,1)
plt.plot(x,y,color='b',linestyle=":",linewidth=4)

plt.subplot(2,1,2)
plt.plot(a,b,color='g',linestyle="--",linewidth=2)

plt.show()
```

```python
# 10. Changing Mark Aesthetics
```

```python
x = [10,20,30,40,50,60,70,80,90]
y = [8,1,7,2,0,3,7,3,2]

plt.scatter(x,y,marker="*",c="r",s=100)
plt.grid(True)
plt.show()
```

```python
# 11. Adding two markers in the same plot
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [4,6,1,9,8,0,5,2,9]

plt.scatter(x,a,marker="*", c="r",s=200)
plt.scatter(x,b,marker=".",c="g",s=100)
plt.show()
```

```python
#12. Adding SubPlots
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [4,6,1,9,8,0,5,2,9]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="r",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=",",c="y",s=100)
plt.show()
```
